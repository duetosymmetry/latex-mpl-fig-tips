```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = .03 + 0.1*np.sin(2. * np.pi * t)
```

```python
def test_plot(filename, figsize=None):
    if figsize is None:
        fig = plt.figure()
    else:
        fig = plt.figure(figsize=figsize)
    plt.plot(t, s)
    plt.xlabel('$t-t_0$')
    plt.ylabel('Strain $h_{22}$')

    plt.tight_layout()

    fig.savefig(f"{filename}.png")
    fig.savefig(f"{filename}.pdf")
    plt.show()
```

```python
test_plot("default_all")
```

```python
mpl.rcParams['text.usetex'] = True
```

```python
test_plot("usetex_True")
```
